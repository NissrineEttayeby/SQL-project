```python
diplome = filiere.nom_filiere[0]
ecole = filiere.nom_école[0]
date_acc = filiere.date_accreditation[0]
```

# Code

```python
#initialisation d'une listes vides qui va contenir toutes les codifications possible HI......
codifications= []

#les tables dont lesquelles on va inserer les données (dans le file_Output.sql)
tab1 ='Mod_Elem_id'
tab2= 'Inscr_Pedag_Id'
tab3= 'Trait_Notes_Id'
tab4= 'List_Mod_Elm_Id'

with open('file_Output.sql','w') as f:
    for AN in range(1,4): #trois ans de cycle d'ingénieur
        codif_1 = 'HI'+diplome.upper()+str(AN)+'04'
        codifications.append(codif_1)
        if AN ==1: 
            f.write(f"--{AN} ère année :\n")
            f.write(f"insert into {tab1} values ('{codif_1}','{ecole}','AN','Etape {AN}',' ','{date_acc}');\n")
        if AN == 2 or AN ==3:
            f.write(f"--{AN} ème année :\n")
            f.write(f"insert into {tab1} values ('{codif_1}','{ecole}','AN','Etape {AN}',' ','{date_acc}');\n")
            
        for semestre in range(1,7): #6 semestres
            if AN==1 and (semestre==1 or semestre==2):
                codif_2 = 'HI'+diplome.upper()+str(semestre)+'004'
                f.write(f"--S{semestre}\n")
                f.write(f"insert into {tab1} values ('{codif_2}','{ecole}','SM0{semestre}','Semestre{semestre}',' ','{date_acc}');\n")
                codifications.append(codif_2)
            elif AN ==2 and (semestre==3 or semestre==4):
                codif_2 = 'HI'+diplome.upper()+str(semestre)+'004'
                f.write(f"--S{semestre}\n")
                f.write(f"insert into {tab1} values ('{codif_2}','{ecole}','SM0{semestre}','Semestre{semestre}',' ','{date_acc}');\n")
                codifications.append(codif_2)
            if AN==3 and (semestre==5 or semestre==6):
                codif_2 = 'HI'+diplome.upper()+str(semestre)+'004'
                f.write(f"--S{semestre}\n")
                f.write(f"insert into {tab1} values ('{codif_2}','{ecole}','SM0{semestre}','Semestre{semestre}',' ','{date_acc}');\n")
                codifications.append(codif_2)
                
            for i in range(len(etude)):
                if etude.nom_mod[i]!=0 and etude.num_SM[i]==semestre and etude.annee[i]==AN:
                    codif_3 = 'HI'+diplome.upper()+str(etude.num_SM[i])+str(etude.MOD[i])+'04'
                    codifications.append(codif_3)
                    f.write(f"insert into {tab1} values ('{codif_3}','{ecole}','MOD','S{semestre}','{etude.nom_mod[i]}',{date_acc});\n")
                    
                if  etude.ELM1[i] != 0 and etude.num_SM[i]==semestre and etude.annee[i]==AN:
                    codif_elm1 = 'HI' + diplome.upper() + str(etude.num_SM[i])+str(etude.MOD[i])+'14'
                    codifications.append(codif_elm1)
                    f.write(f"insert into {tab1} values ('{codif_elm1}','{ecole}','ELM','S{semestre}','{etude.ELM1[i]}',{date_acc});\n")
                    
                if etude.ELM2[i] != 0 and etude.num_SM[i]==semestre and etude.annee[i]==AN:
                    codif_elm2 = 'HI'+diplome.upper()+str(etude.num_SM[i])+str(etude.MOD[i])+'24'
                    codifications.append(codif_elm2)
                    f.write(f"insert into {tab1} values ('{codif_elm2}','{ecole}','ELM','S{semestre}','{etude.ELM2[i]}',{date_acc});\n") 
                    
                if etude.ELM3[i] != 0 and etude.num_SM[i]==semestre and etude.annee[i]==AN:
                    codif_elm3 = 'HI'+diplome.upper()+str(etude.num_SM[i])+str(etude.MOD[i])+'34'
                    codifications.append(codif_elm3)
                    f.write(f"insert into {tab1} values ('{codif_elm3}','{ecole}','ELM','S{semestre}','{etude.ELM3[i]}',{date_acc});\n")
                    
                if etude.ELM4[i] != 0 and etude.num_SM[i]==semestre and etude.annee[i]==AN:
                    codif_elm4 = 'HI'+diplome.upper()+str(etude.num_SM[i])+str(etude.MOD[i])+'44'
                    codifications.append(codif_elm4)
                    f.write(f"insert into {tab1} values ('{codif_elm4}','{ecole}','ELM','S{semestre}','{etude.ELM4[i]}',{date_acc});\n")
                    
                #if etude.nom_mod[i]==0 and etude.num_SM[i]==semestre and etude.annee[i]==AN:
                    #codif_4 = 'HI'+diplome.upper()+str(etude.num_SM[i])+str(etude.MOD[i])+'04'
                    #f.write(f"insert into {tab1} values ('{codif_4}','{ecole}','MOD','S{semestre}','no module exist',{date_acc});\n")
                    #codifications.append(codif_4)
        f.write(f"\n")
    f.write("commit;\n")
    f.write(f"\n")
    f.write(f"\n")
    f.write(f"\n")
    
    #Insertion dans la table : 'Inscr_Pedag_Id'
    
    for cod in codifications:
        f.write(f"insert into {tab2} values ('{ecole}',{cod});\n")
       
    f.write("commit;\n")
    f.write(f"\n")
    f.write(f"\n")
    
    #Insertion dans la table : 'Trait_Notes_Id'
    
    for cod in codifications:
        f.write(f"insert into {tab3} values ({cod},'HTN','T','');\n")
        
    f.write("commit;\n")
    f.write(f"\n")
    f.write(f"\n")
    
    
    #Insertion dans la table : 'List_Mod_Elm_Id'
    
    for AN in range(1,4): #trois ans de cycle d'ingénieur
        codif_0 = 'HI'+diplome.upper()+str(AN)+'4'
        f.write(f"insert into {tab4} values ('{codif_0}','O','O',' VET {AN} Année ID','');\n")
        codif_1 = 'HI'+diplome.upper()+str(AN)+'04'
        #codifications.append(codif_1)
        f.write(f"insert into {tab4} values ('{codif_1}','O','O','VET_ELEM_NOM','','');\n")

        for semestre in range(1,7): #6 Semestres
            if AN==1 and (semestre==1 or semestre==2):
                codif_2 = 'HI'+diplome.upper()+str(semestre)+'004'
                f.write(f"insert into {tab4} values ('{codif_2}','O','O',Semestre{semestre}',' ','');\n")
                #codifications.append(codif_2)
                
            elif AN ==2 and (semestre==3 or semestre==4):
                codif_2 = 'HI'+diplome.upper()+str(semestre)+'004'
                f.write(f"insert into {tab4} values ('{codif_2}','O','O',Semestre{semestre}',' ','');\n")
                #codifications.append(codif_2)
                
            if AN==3 and (semestre==5 or semestre==6):
                codif_2 = 'HI'+diplome.upper()+str(semestre)+'004'
                f.write(f"insert into {tab4} values ('{codif_2}','O','O',Semestre{semestre}',' ','');\n")
                #codifications.append(codif_2)
                
            for i in range(len(etude)):
                if etude.nom_mod[i]!=0 and etude.num_SM[i]==semestre and etude.annee[i]==AN:
                    codif_3 = 'HI'+diplome.upper()+str(etude.num_SM[i])+str(etude.MOD[i])+'04'
                    #codifications.append(codif_3)
                    f.write(f"insert into {tab4} values ('{codif_3}','O','O','{etude.nom_mod[i]}','');\n") 
                    
                #if etude.nom_mod[i]==0 and etude.num_SM[i]==semestre and etude.annee[i]==AN:
                    #codif_4 = 'HI'+diplome.upper()+str(etude.num_SM[i])+str(etude.MOD[i])+'04'
                    #f.write(f"insert into {tab4} values ('{codif_4}','O','O','no module exist','');\n")
                    #codifications.append(codif_4)
                    
        f.write(f"\n")
    f.write("commit;\n")
    f.write(f"\n")
    f.write(f"\n")
    f.write(f"\n")    
```
